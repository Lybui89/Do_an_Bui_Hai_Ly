```python
from tkinter import *
from tkinter import ttk
from tkinter import messagebox
import json

DATA_FILE = "data_nuochoa.json"
mang_nuoc_hoa = [] #tương ứng với nội dung file data_hanghoa.json

root = Tk()
root.title("Quản lý Nước hoa")
root.geometry("1000x400")


# Chia FRAME
list_nuochoa_frame = ttk.Frame(root)
list_button_frame = ttk.Frame(root)
input_nuochoa_frame = ttk.Frame(root)
list_nuochoa_frame.grid(row=0, column=0)
list_button_frame.grid(row=1, column=0)
input_nuochoa_frame.grid(row=2, column=0)

# Chèn treeview
# Ds cột (key)
columns=('ma_nuochoa', 'ten_nuochoa', 'gia_ban', 'so_luong', 'loai')
tree = ttk.Treeview(list_nuochoa_frame, columns=columns, show='headings')
tree.heading('ma_nuochoa', text='Mã nước hoa')
tree.heading('ten_nuochoa', text='Tên nước hoa')
tree.heading('gia_ban', text='Đơn giá bán')
tree.heading('so_luong', text='Số lượng')
tree.heading('loai', text='Chủng loại')
tree.pack()


def doc_du_lieu():
    '''Đọc dữ liệu từ file data_nuochoa.json và gán cho biến mảng nước hoa dạng list []'''
    global mang_nuoc_hoa
    try:
        with open(DATA_FILE, "r", encoding="utf-8") as myfile:            
            mang_nuoc_hoa = json.load(myfile)
    except:
        messagebox.showwarning(title="Lỗi", message="Lỗi đọc file")
        mang_nuoc_hoa = []

        
def luu_du_lieu():
    '''Lưu dữ liệu hàng hóa từ biến mảng các hàng hóa xuống file data_nuochoa.json'''
    try:
        with open(DATA_FILE, "w", encoding="utf-8") as myfile:
            json.dump(mang_nuoc_hoa, myfile, indent = 4)
    except:
        messagebox.showwarning(title="Lỗi", message="Lỗi lưu file")

        
def tim_nuoc_hoa_theo_ma(id):
    """Tìm nước hoa dựa vào mã nước hoa, trả về hàng hóa tìm được dạng dict hoặc None."""
    for item in mang_nuoc_hoa:
        if item["ma_nuochoa"] == id:
            return item
    
    return None


def hien_thi_danh_sach_nuoc_hoa():
    """Cập nhật lưới"""
    clear_tree_view(tree)
    if len(mang_nuoc_hoa) == 0:
        doc_du_lieu()        
    for item in mang_nuoc_hoa:
        tree.insert(
            parent='', index='end',
            values=(item['ma_nuochoa'], item['ten_nuochoa'], item['gia_ban'], item['so_luong'], item['loai'])
        )
            
def clear_tree_view(mytree):
    for item in mytree.get_children():
        mytree.delete(item)


def them_nuoc_hoa():
    #TODO: Check nhập liệu hợp lệ
    obj = {
        "ma_nuochoa": manuochoa_value.get(),
        "ten_nuochoa": tennuochoa_value.get(),
        "gia_ban": dongia_value.get(),
        "so_luong": so_luong_value.get(),
        "loai": gia_tri_chon_loai.get()
    }
    global mang_nuoc_hoa
    
    # Kiểm tra xem mã hàng hóa đã có hay chưa?
    nuoc_hoa_tim = tim_nuoc_hoa_theo_ma(obj['ma_nuochoa'])
    
    if nuoc_hoa_tim is not None:
        messagebox.showwarning(message=f"Đã có nước hoa mã: {obj['ma_nuochoa']}")
    else:    
        # Nếu chưa thêm ==> cập nhật treeview
        mang_nuoc_hoa.append(obj)
        # Save file
        luu_du_lieu()
        # Cập nhật treeview
        hien_thi_danh_sach_nuoc_hoa()
        messagebox.showinfo(message="Thêm mới thành công!")

# Định nghĩa hàm xử lý sự kiện chọn và gán
def xu_ly_chon_nuoc_hoa(e):
    for i_item in tree.selection():
        selected_item = tree.item(i_item)
        row_value = selected_item["values"]
        
        # Cập nhật thông tin xuống dưới (vùng thông tin)
        manuochoa_value.set(int(row_value[0]))
        tennuochoa_value.set(row_value[1])
        dongia_value.set(row_value[2])
        so_luong_value.set(row_value[3])
        gia_tri_chon_loai.set(row_value[4])

        
def xoa_nuoc_hoa(manuochoa):
    # 1.Tìm hàng hóa theo mahh
    nuoc_hoa_tim_thay = tim_nuoc_hoa_theo_ma(manuochoa)
    if nuoc_hoa_tim_thay is not None:
        # 2. Nếu có thì
        # 2.1 xóa khỏi mang_nuoc_hoa
        global mang_nuoc_hoa
        mang_nuoc_hoa.remove(nuoc_hoa_tim_thay)
        # 2.2 update treeview
        hien_thi_danh_sach_nuoc_hoa()
        # 2.3 lưu xuống file
        luu_du_lieu()
    
        
def rightClickMenu(event):
    # create a popup menu
    rowID = tree.identify('item', event.x, event.y)
    if rowID:
        tree.selection_set(rowID)
        tree.focus_set()
        tree.focus(rowID)
        menu = Menu(tree, tearoff=0)
        menu.add_command(label="Delete", command=lambda: xoa_nuoc_hoa(manuochoa_value.get()))
        menu.add_command(label="Clear All", command= lambda: clear_tree_view(tree))
        menu.post(event.x_root, event.y_root)
    else:
        pass

tree.bind('<<TreeviewSelect>>', xu_ly_chon_nuoc_hoa)
tree.bind('<3>', rightClickMenu)


def xoa_thong_tin_nuoc_hoa():
    manuochoa_value.set(0)
    tennuochoa_value.set('')
    dongia_value.set(0)
    so_luong_value.set(0)
    gia_tri_chon_loai.set('')


# Dinh nghia widget cho list_button_frame
Button(list_button_frame, text="Xóa danh sách",
    command= lambda: clear_tree_view(tree)
).grid(row=0, column=0, padx=5, pady=5)

Button(list_button_frame, text="Cập nhật danh sách",
       command= lambda: hien_thi_danh_sach_nuoc_hoa()
).grid(row=0, column=1, padx=5, pady=5)

Button(list_button_frame, text="Thêm", command=them_nuoc_hoa).grid(row=0, column=2, padx=5, pady=5)
Button(list_button_frame, text="Xóa", command=xoa_thong_tin_nuoc_hoa).grid(row=0, column=3, padx=5, pady=5)
Button(list_button_frame, text="Cập nhật").grid(row=0, column=4, padx=5, pady=5)

# Lấy danh sách nước hoa lúc form chạy
doc_du_lieu()
hien_thi_danh_sach_nuoc_hoa()


# Thêm widget cho vùng input
Label(input_nuochoa_frame, text="Mã nước hoa").grid(row=0, column=0)
manuochoa_value = IntVar()
manuochoa = Entry(input_nuochoa_frame, textvariable=manuochoa_value)
manuochoa.grid(row=0, column=1)

Label(input_nuochoa_frame, text="Tên hàng hóa").grid(row=1, column=0)
tennuochoa_value = StringVar()
tennuochoa = Entry(input_nuochoa_frame, textvariable=tennuochoa_value)
tennuochoa.grid(row=1, column=1)

Label(input_nuochoa_frame, text="Loại").grid(row=2, column=0)
gia_tri_chon_loai = StringVar()
loai = ttk.Combobox(input_nuochoa_frame, textvariable=gia_tri_chon_loai)
loai.grid(row=2, column=1)
loai['values'] = ["Eau de Perfum", "Eau de Toilette"]


Label(input_nuochoa_frame, text="Đơn giá").grid(row=3, column=0)
dongia_value = IntVar()
dongia = Entry(input_nuochoa_frame, textvariable=dongia_value)
dongia.grid(row=3, column=1)

Label(input_nuochoa_frame, text="Số lượng").grid(row=4, column=0)
so_luong_value = IntVar()
soluong = Entry(input_nuochoa_frame, textvariable=so_luong_value)
soluong.grid(row=4, column=1)

root.mainloop()
```
